## III.A 5-fold Cross Validation

The idea behind 5-fold cross-validation or k-fold cross validation is to divide all the available data items into roughly equal-sized sets. Each set is used exactly once as the test set while the remaining data is used as the training set. The idea of cross validation is to choose the representation in which the error of the validation set is a minimum. In these cases, learning can continue until the error of the validation set starts to increase.

This is used to:<br>
-> Estimate skill of model on new data.<br>
-> It is a popular method because it is simple to understand and because it generally results in a less biased or less optimistic estimate of the model skill than other methods, such as a simple train/test split.<br>

Following are the steps to perform k-fold cross validation:

1) We need to shuffle the dataset initially in random order.<br>
2) Then, we need to split the data into k groups, in our assignment it is 5.<br>
3) For each unique group:

    ->Take the group as a hold out or test data set<br>
    ->Take the remaining groups as a training data set<br>
    ->Fit a model on the training set and evaluate it on the test set<br>
    Retain the evaluation score and discard the model<br>
4) Summarize the skill of the model using the sample of model evaluation scores<br>

<H2> Nonlinear Regression using Neural Networks</H2>

A type of Logistic Regression that incorporates forward pass propagation and backwards backpropagation, unlike the Linear Logistic Regression.

```python

""" Neural Network 
        referenced NN code by Chuck Anderson in R and C++ 

                                by Jake Lee (lemin)

    example usage:
        X = numpy.array([0,0,1,0,0,1,1,1]).reshape(4,2)
        T = numpy.array([0,1,1,0,1,0,0,1]).reshape(4,2)

        nn = nnet.NeuralNet([2,3,2])
        nn.train(X,T, wprecision=1e-20, fprecision=1e-2)
        Y = nn.use(X)

"""
from grad import scg, steepest
from copy import copy


class NeuralNet:
    """ neural network class for regression
        
        Parameters
        ----------
        nunits: list
            the number of inputs, hidden units, and outputs

        Methods
        -------
        set_hunit    
            update/initiate weights

        pack 
            pack multiple weights of each layer into one vector

        forward
            forward processing of neural network

        backward
            back-propagation of neural network

        train
            train the neural network

        use
            appply the trained network for prediction

        Attributes
        ----------
        _nLayers
            the number of hidden unit layers 

        rho
            learning rate

        _W
            weights
        _weights
            weights in one dimension (_W is referencing _weight)

        stdX
            standardization class for data
        stdT
            standardization class for target

        Notes
        -----
        
    """

    def __init__(self, nunits):

        self._nLayers=len(nunits)-1
        self.rho = [1] * self._nLayers
        self._W = []
        wdims = []
        lenweights = 0
        for i in range(self._nLayers):
            nwr = nunits[i] + 1
            nwc = nunits[i+1]
            wdims.append((nwr, nwc))
            lenweights = lenweights + nwr * nwc

        self._weights = np.random.uniform(-0.1,0.1, lenweights) 
        start = 0  # fixed index error 20110107
        for i in range(self._nLayers):
            end = start + wdims[i][0] * wdims[i][1] 
            self._W.append(self._weights[start:end])
            self._W[i].resize(wdims[i])
            start = end

        self.stdX = None
        self.stdT = None
        self.stdTarget = True

    def add_ones(self, w):
        return np.hstack((np.ones((w.shape[0], 1)), w))

    def get_nlayers(self):
        return self._nLayers

    def set_hunit(self, w):
        for i in range(self._nLayers-1):
            if w[i].shape != self._W[i].shape:
                print("set_hunit: shapes do not match!")
                break
            else:
                self._W[i][:] = w[i][:]

    def pack(self, w):
        return np.hstack(map(np.ravel, w))

    def unpack(self, weights):
        self._weights[:] = weights[:]  # unpack

    def cp_weight(self):
        return copy(self._weights)

    def RBF(self, X, m=None,s=None):
        if m is None: m = np.mean(X)
        if s is None: s = 2 #np.std(X)
        r = 1. / (np.sqrt(2*np.pi)* s)  
        return r * np.exp(-(X - m) ** 2 / (2 * s ** 2))

    def forward(self,X):
        t = X 
        Z = []

        for i in range(self._nLayers):
            Z.append(t) 
            if i == self._nLayers - 1:
                t = np.dot(self.add_ones(t), self._W[i])
            else:
                t = np.tanh(np.dot(self.add_ones(t), self._W[i]))
                #t = self.RBF(np.dot(np.hstack((np.ones((t.shape[0],1)),t)),self._W[i]))
        return (t, Z)
        
    def backward(self, error, Z, T, lmb=0):
        delta = error
        N = T.size
        dws = []
        for i in range(self._nLayers - 1, -1, -1):
            rh = float(self.rho[i]) / N
            if i==0:
                lmbterm = 0
            else:
                lmbterm = lmb * np.vstack((np.zeros((1, self._W[i].shape[1])),
                            self._W[i][1:,]))
            dws.insert(0,(-rh * np.dot(self.add_ones(Z[i]).T, delta) + lmbterm))
            if i != 0:
                delta = np.dot(delta, self._W[i][1:, :].T) * (1 - Z[i]**2)
        return self.pack(dws)

    def _errorf(self, T, Y):
        return T - Y
        
    def _objectf(self, T, Y, wpenalty):
        return 0.5 * np.mean(np.square(T - Y)) + wpenalty

    def train(self, X, T, **params):

        verbose = params.pop('verbose', False)
        # training parameters
        _lambda = params.pop('Lambda', 0.)

        #parameters for scg
        niter = params.pop('niter', 1000)
        wprecision = params.pop('wprecision', 1e-10)
        fprecision = params.pop('fprecision', 1e-10)
        wtracep = params.pop('wtracep', False)
        ftracep = params.pop('ftracep', False)

        # optimization
        optim = params.pop('optim', 'scg')

        if self.stdX == None:
            explore = params.pop('explore', False)
            self.stdX = Standardizer(X, explore)
        Xs = self.stdX.standardize(X)
        if self.stdT == None and self.stdTarget:
            self.stdT = Standardizer(T)
            T = self.stdT.standardize(T)
        
        def gradientf(weights):
            self.unpack(weights)
            Y,Z = self.forward(Xs)
            error = self._errorf(T, Y)
            return self.backward(error, Z, T, _lambda)
            
        def optimtargetf(weights):
            """ optimization target function : MSE 
            """
            self.unpack(weights)
            #self._weights[:] = weights[:]  # unpack
            Y,_ = self.forward(Xs)
            Wnb=np.array([])
            for i in range(self._nLayers):
                if len(Wnb)==0: Wnb=self._W[i][1:,].reshape(self._W[i].size-self._W[i][0,].size,1)
                else: Wnb = np.vstack((Wnb,self._W[i][1:,].reshape(self._W[i].size-self._W[i][0,].size,1)))
            wpenalty = _lambda * np.dot(Wnb.flat ,Wnb.flat)
            return self._objectf(T, Y, wpenalty)

        if optim == 'scg':
            result = scg(self.cp_weight(), gradientf, optimtargetf,
                                        wPrecision=wprecision, fPrecision=fprecision, 
                                        nIterations=niter,
                                        wtracep=wtracep, ftracep=ftracep,
                                        verbose=False)
            self.unpack(result['w'][:])
            self.f = result['f']
        elif optim == 'steepest':
            result = steepest(self.cp_weight(), gradientf, optimtargetf,
                                nIterations=niter,
                                xPrecision=wprecision, fPrecision=fprecision,
                                xtracep=wtracep, ftracep=ftracep )
            self.unpack(result['w'][:])
        if ftracep:
            self.ftrace = result['ftrace']
        if 'reason' in result.keys() and verbose:
            print(result['reason'])

        return result

    def use(self, X, retZ=False):
        if self.stdX:
            Xs = self.stdX.standardize(X)
        else:
            Xs = X
        Y, Z = self.forward(Xs)
        if self.stdT is not None:
            Y = self.stdT.unstandardize(Y)
        if retZ:
            return Y, Z
        return Y
```

Class defined above is Neural net class.<br>

<li>This class builds the basic neural network which would learn from forward and backward propagation.<br> 

<li>Parameter nunits takes the number of inputs, hidden units, and outputs which are needed for the model to run <br>

<li>def set_hunit(self, w) is used to update the weights. w parameter is the weight.<br>

<li>def pack(self, w) is used to pack the weights. <br>

<li>Forward Pass:
$$
Z = h(X1 \cdot V) \\
Y  = Z1 \cdot W \\
G  = softmax(Y)
$$

<br>

<li>Backward Pass:
$$
\begin{align}
V &\leftarrow V + \alpha_h X1^\top \Big( (T - G) W^\top \odot (1 - Z^2) \Big)\\
W &\leftarrow W + \alpha_o Z1^\top \Big(  T - G\Big) 
\end{align}
$$

<br>
    
<li>Fitting the data with MLP is a combinatorial optimization problem with non-smooth step function. 
So, we can consider smooth step function, a s-shaped sigmoid function. 
We call this smooth function as **activation function**. <br>

<li>The backward() method is used to update weights to minimize the cost. This method  updates weights during back propagation. The objective function is the cross entropy function and we aim to maximise the likelihood of the samples belonging to the class. We can use the formulas mentioned in the nonlinear classification model to compute forward(), backward() and objective function _objectf() i.e error cost function .  <br>

<li>The train() method  converts the target variable as an indicator variable. From this method, the forward(), backward() and _objectf() methods are called to update weights and to maximise the likelihood of a sample belonging to a class. <br>

<li>The use() method calls the forward() method to get the probabilities and the argmax() method is used on these probabilities to get the class with maximum probability as the output for the given samples. <br>


```python
def partition(X, T, ratio): 
    
    assert(np.sum(ratio) == 1)
    
    # shuffle the data indices 
    idxs = np.random.permutation(X.index)
    
    # the number of samples 
    N = X.shape[0]
    
    Xs = []
    Ts = []
    i = 0  # first index to zero
    for k, r in enumerate(ratio):
        nrows = int(round(N * r))  # number of rows
        
        if k == len(ratio) -1:
            Xs.append(X.iloc[i:, :])
            Ts.append(T.iloc[i:, :])
        else:
            Xs.append(X.iloc[i:i+nrows, :])
            Ts.append(T.iloc[i:i+nrows, :])
        
        i += nrows
    
    return Xs, Ts
```

```python
import pandas as pd
##params contains learning rate and number of hidden units in a layer
def KFold_cross_validation_regression(X,T,K,params):
    output_units= T.shape[1]
    input_units= X.shape[1]
    print("input features:"+str(input_units))
    for test in range(K):
        ratio=[0.2,0.8]
        Xlst, Tlst = partition(pd.DataFrame(X), pd.DataFrame(T),ratio)
        Xtest, X_rem = [xx.to_numpy() for xx in Xlst]
        Ttest, T_rem = [tt.to_numpy() for tt in Tlst]
        param_size=len(params)  
        records=[]
        for val in range(K):
            par_rat =[0.2,0.8]
            X_l, T_l = partition(pd.DataFrame(X_rem), pd.DataFrame(T_rem),par_rat)    
            Xval,Xtrain= [xx.to_numpy() for xx in X_l]
            Tval,Ttrain= [xx.to_numpy() for xx in T_l]
             
            rec=np.zeros(shape=(100,100))
            
            for param in params:
                input_lis=[]
                input_lis.append(input_units)
                input_lis.extend(param)
                input_lis.append(output_units)
                #print(str(input_lis))
                nn = NeuralNet(input_lis)
                nn.train(Xtrain, Ttrain,ftracep=True)
                pred,z = nn.use(Xval, retZ=True)
                #print("Tval: "+str(Tval.shape))
                #print("pred: "+str(pred.shape))
                records.append(np.sqrt(np.mean((Tval- pred)**2))) ##RMSE
                
        # records of valAccs
        #print(records)
        bestParam=params[np.argmin(records)%param_size]
        final_lis=[]
        final_lis.append(input_units)
        final_lis.extend(bestParam)
        final_lis.append(output_units)
        bestModel = NeuralNet(final_lis)
        X_train=X_rem
        T_train=T_rem
        bestModel.train(X_train, T_train,ftracep=True)
        finalPred,prob = bestModel.use(Xtest, retZ=True)
        final_rmse = np.sqrt(np.mean((Ttest- finalPred)**2))
        print("RMSE: "+str(final_rmse)+" params: "+ str(bestParam))
```

The KFold_cross validation method takes the input data, target variable, "K" representing the number of folds and "params" representing the different parameters (learning rate, number of hidden units in each hidden layer) for neural networks as argument.

The partition() method is used to divide the dataset based on the given ratio.

The K-fold cross validation is used to obtain the best parameters to train my model for the data. 

Firstly, I use the partition() method to partition my data into 80% and 20% of my samples. Then, I use that 80% data and again call partition() method to partition it into 80% train and 20% validation set. 

The initial 20% of data is used as the first test set. The parameter to my cross validation is learning rate and the hidden units in each hidden layer ([[10, [10]],[0.1,[20]],[10,[20]],[0.1,[20,1,2]]). 

I train the model using all the parameters and record the accuracy of each parameter. 

Then, I repeat the same steps for "K" times with different train and validation sets. 

I choose the best parameter giving the highest accuracy to train my model and then use the trained model on my test set. 

Then, I repeat this "K" times for different test sets each involving "K" executions of training and validation part. 

Finally, I output the best parameter for each of my test set in "K" executions and choose the parameter with highest accuracy and that occurs multiple times.

<H2>Nonlinear Logistic Regression</H2>

We can extend the linear logistic regression model to achieve nonlinear classification using neural networks. We can use two layer neural networks to solve machine learning problems.

By using this softmax function, we were able to generate probablistic outputs for all classes. To handle multi-label classes, we use the indicator target labels for training to update the weights.<br>

Softmax Function:
$$
g_k(x) = P(T=k \mid x) = \frac{e^{\kappa_k}}{\sum_{c=1}^K e^{\kappa_c}}
$$

Forward Pass:
$$
Z = h(X1 \cdot V) \\
Y  = Z1 \cdot W \\
G  = softmax(Y)
$$

Backward Pass:
$$
\begin{align}
V &\leftarrow V + \alpha_h X1^\top \Big( (T - G) W^\top \odot (1 - Z^2) \Big)\\
W &\leftarrow W + \alpha_o Z1^\top \Big(  T - G\Big) 
\end{align}
$$

Since we maximize the likelihood function, we define our error function as the negative logarithm of it:

$$
E(w) = - \ln P(T \mid w) =  - \sum_{n=1}^{N} \sum_{k=1}^{K} t_{n,k} \ln y_{n,k}.
$$
This function is called cross-entropy error function for the multiclass classification problem.

The <b> NeuralNetLogReg </b> class is the subclass of NeuralNet class. It inherits all the functions in NeuralNet class and also the variables in the NeuralNet class. We can reuse the variables and override the methods of NeuralNet class in NeuralNetLogReg class to solve regression problem and create a nonlinear regression model using neural networks.

The init() method in the NeuralNetLogReg class is used to initialize the essential variables for the NeuralNetLogReg class. We initialize all the variables by calling the super class init() method and additionally change few variables to support our regression. We add variable "n_classes" to keep track of number of classes in the target variable. The learning rate parameter is also modified to support different learning rate inputs. The variable "stdTarget" is made false , so that we do not standardize the target variable in our classification.

The forward() method calls the superclass method to compute the values at each layer in the forward pass and we finally use the softmax function on it to get the probability values.

The backward() method is used to update weights to minimize the cost. This method is reused from the NeuralNet class and updates weights during back propagation. The objective function is the cross entropy function and we aim to maximise the likelihood of the samples belonging to the class. We can use the formulas mentioned in the nonlinear classification model to compute forward(), backward() and objective function _objectf() i.e error cost function .

The train() method can be reused from the super class NeuralNet after converting the target variable as an indicator variable. From this method, the forward(), backward() and _objectf() methods are called to update weights and to maximise the likelihood of a sample belonging to a class.

The use() method calls the forward() method to get the probabilities and the argmax() method is used on these  probabilities  to get the class with maximum probability as the output for the given samples.


```python
from nn import NeuralNet
from util import Standardizer
from grad import scg
from grad import steepest
import sys
import warnings

if not sys.warnoptions:
    warnings.simplefilter("ignore")
    
class NeuralNetLogReg(NeuralNet):
    """ Nonlinear Logistic Regression
    """

    
    def __init__(self, nunits,alpha):
        NeuralNet.__init__(self,nunits)
        self.n_classes=None
        self.rho = [1*alpha] * self._nLayers
        self.stdTarget = False
        
    def softmax(self,z):
        if not isinstance(z, np.ndarray):
            z = np.asarray(z)
            
        f = np.exp(np.longfloat(z))
        return f / (np.sum(f, axis=1, keepdims=True) if len(z.shape) == 2 else np.sum(f))
      
  
    def forward(self, X):
        t,Z= super(NeuralNetLogReg,self).forward(X)
        G=self.softmax(t)
        return (G, Z)
    
      
    def _objectf(self, T, Y, wpenalty):
        return - (np.sum(np.multiply(T, np.log(Y))))+wpenalty
                        
                       
    def train(self, X, T, **params):
        self.n_classes=np.unique(T)     
        ##converting T to indicator variable
        N=X.shape[0]
        #print(np.unique(T))
        self.n_classes=np.unique(T)
        T1 = np.zeros((N, len(self.n_classes)))
        i=0
        for c in self.n_classes:
            idx=np.where((T==c)==True)[0]
            T1[idx,i] =1
            i=i+1
        T = T1
        super(NeuralNetLogReg,self).train(X,T,**params)
            
    def use(self, X):
        if self.stdX:
            Xs = self.stdX.standardize(X)
        else:
            Xs = X
        Y, Z = self.forward(Xs)
        y = np.argmax(Y,axis=1)
        return self.n_classes[y],Y
```
